```python
import pandas as pd


# Dataset ini tidak punya header.
# Nilai seperti 2001 di kolom pertama adalah TARGET (tahun), bukan nama feature /  kolom.
file_path = '/content/midterm-regresi-dataset.csv'
df = pd.read_csv(file_path)

df.head()
```

```python
print("df.shape:", df.shape) #ada 515344 row/baris dan 90 feature
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

print("ðŸš€ Mulai Preprocessing Data Regresi...")

# 1. Pisahkan Fitur (X) dan Target (y)
# Kolom pertama (index 0) adalah 'Tahun' (Target yang mau ditebak)
# Kolom sisanya (index 1 sampai 90) adalah fitur audio
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values

# 2. Scaling (Standarisasi)
# Ini WAJIB untuk Regresi agar angka fitur (timbre) yang variatif jadi setara
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Split Data (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("âœ… Data Siap!")
print(f"Jumlah Data Latih: {X_train.shape[0]}")
print(f"Jumlah Data Uji  : {X_test.shape[0]}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("ðŸ¤– Training Model Machine Learning (Linear Regression)...")

# 1. Inisialisasi & Latih Model
ml_model = LinearRegression()
ml_model.fit(X_train, y_train)

# 2. Prediksi ke Data Test
y_pred = ml_model.predict(X_test)

# 3. Evaluasi Hasil
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nðŸ“Š Hasil Evaluasi Machine Learning:")
print(f"MAE (Rata-rata Meleset): {mae:.2f} Tahun")
print(f"R2 Score: {r2:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

print("ðŸ”§ Mulai Hyperparameter Tuning (Pakai Random Forest)...")

# Kita ganti modelnya jadi Random Forest
# Hyperparameter yang kita tuning (setting):
# 1. n_estimators=100 -> Kita pakai 100 "pohon" keputusan (makin banyak makin akurat, tapi lama)
# 2. max_depth=20     -> Kedalaman pohon dibatasi 20 biar gak menghapal data (overfitting)
# 3. random_state=42  -> Biar hasilnya konsisten
rf_model = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42, n_jobs=-1)

# Latih Model
rf_model.fit(X_train, y_train)

# Prediksi & Evaluasi
y_pred_rf = rf_model.predict(X_test)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("\nðŸ“Š Hasil Tuning (Random Forest):")
print(f"MAE (Rata-rata Meleset): {mae_rf:.2f} Tahun")
print(f"R2 Score: {r2_rf:.4f}")

# Bandingkan dengan Linear Regression tadi
print("\n------------------------------------------------")
print(f"Perbandingan:")
print(f"1. Linear Regression : Error {mae:.2f} Tahun")
print(f"2. Random Forest     : Error {mae_rf:.2f} Tahun")

if mae_rf < mae:
    print("âœ… Tuning Berhasil! Error turun.")
else:
    print("âŒ Hmm, tidak ada perubahan signifikan.")
```
